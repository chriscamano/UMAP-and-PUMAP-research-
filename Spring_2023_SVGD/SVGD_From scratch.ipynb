The core idea of SVGD involves approximating a target distribution, p, through an approximate distribution, q. This is achieved by performing a series of transformations on q, which gradually move it towards p.

In order to build up to Stien Variational Gradient Descent and its implementation we will need working knowledge of a few key concepts

## Reproducing Kernel Hilbert Spaces

Let $$k\left(x, x^{\prime}\right): \mathcal{X} \times \mathcal{X} \rightarrow \mathbb{R}$$
be a positive definite kernel. The reproducing kernel Hilbert space (RKHS) $\mathcal{H}$ of $k\left(x, x^{\prime}\right)$ is the closure of linear span 

$$\left\{f: f(x)=\sum_{i=1}^m a_i k\left(x, x_i\right), a_i \in \mathbb{R}, m \in \mathbb{N}, x_i \in \mathcal{X}\right\}$$, equipped with inner products 

$$\langle f, g\rangle_{\mathcal{H}}=\sum_{i j} a_i b_j k\left(x_i, x_j\right)$$ 

for $g(x)=\sum_i b_i k\left(x, x_i\right)$. Denote by $\mathcal{H}^d$ the space of vector functions $\boldsymbol{f}=\left[f_1, \ldots, f_d\right]$ with $f_i \in \mathcal{H}$, equipped with inner product

$$\langle\boldsymbol{f}, \boldsymbol{g}\rangle_{\mathcal{H}^d}=\sum_{i=1}^d\left\langle f_i, g_i\right\rangle_{\mathcal{H}}$$. 

This space is very important in this optimization problem and will be used to alter our perspective of the next concept needed for building this algorithm, the Stien Identity and Stien Discrepancy 


## The Stien Identity and Stien Discrepancy 

Stein's Identity is the key theorem we use in this algorithm to derive the necessary criteria for performing optimizations during the algorithm.  Let $p(x)$ be a continuously differentiable density supported on $\mathcal{X} \subseteq \mathbb{R}^d$, and $\phi(x)=\left[\phi_1(x), \cdots, \phi_d(x)\right]^{\top}$ be a smooth vector function. Stein's identity states that for sufficiently regular $\phi$, we have
$$
\mathbb{E}_{x \sim p}\left[\mathcal{A}_p \boldsymbol{\phi}(x)\right]=0, \quad \text { where } \quad \mathcal{A}_p \phi(x)=\boldsymbol{\phi}(x) \nabla_x \log p(x)^{\top}+\nabla_x \boldsymbol{\phi}(x),
$$

where $\mathcal{A}_p$ is called the Stein operator, which acts on function $\phi$ and yields a zero mean function $\mathcal{A}_p \phi(x)$ under $x \sim p$. 

Now let $q(x)$ be a different smooth density also supported in $\mathcal{X}$, and consider the expectation of $\mathcal{A}_p \phi(x)$ under $x \sim q$, then $\mathbb{E}_{x \sim q}\left[\mathcal{A}_p \phi(x)\right]$ would no longer equal zero for general $\phi$. Instead, the magnitude of $\mathbb{E}_{x \sim q}\left[\mathcal{A}_p \phi(x)\right]$ relates to how different $p$ and $q$ are, and can be leveraged to define a discrepancy measure, known as Stein discrepancy, by considering the "maximum violation of Stein's identity" for $\phi$ in some proper function set $\mathcal{F}$ :
$$
\mathbb{S}(q, p)=\max _{\phi \in \mathcal{F}}\left\{\left[\mathbb{E}_{x \sim q} \operatorname{trace}\left(\mathcal{A}_p \phi(x)\right)\right]^2\right\}
$$

We can determine essentially how "different" two functions are. A very useful property to have when working on approximating distributions in a bayesian learning context. 

## The Kernalized Stien Discrepancy 

The proper function set $\mathcal{F}$ mentioned earlier is often difficult or impossible to optimize over so instead we can constrain this search to the resulting reproducing kernel hilbert space of a function $\phi$ to reframe the problem as 

$$\mathbb{S}(q, p)=\max _{\phi \in \mathcal{H}^d}\left\{\left[\mathbb{E}_{x \sim q}\left(\operatorname{trace}\left(\mathcal{A}_p \phi(x)\right)\right)\right]^2, \quad \text { s.t. } \quad\|\phi\|_{\mathcal{H}^d} \leq 1\right\}$$

which has a known solution from previous research which was found to be : 

$$\phi(x)=\frac{\phi_{q, p}^*(x) }{\left\|\phi_{q, p}^*\right\|_{\mathcal{H}^d}}
$$

which in turn gives 

$$
\phi_{q, p}^*(\cdot)=\mathbb{E}_{x \sim q}\left[\mathcal{A}_p k(x, \cdot)\right]
$$

## Framing this information from the perspective of gradient descent 

Suppose we have an initial distribution $q$, which we pass through a transformation $T: \mathbb{R}^N \rightarrow \mathbb{R}^N$, that is $z=T(x), \quad$ where $x \sim q(x)$
If the map $T$ is invertible, we can easily compute the density of the transformed variable $z$ via the change of variables formula. To ensure $T$ is invertible, let us set $$T(x)=x+\epsilon \phi(x)$$, where $\epsilon$ is a small coefficient and $\phi: \mathbb{R}^N \rightarrow \mathbb{R}^N$. If $\phi$ is smooth and $\epsilon$ is sufficiently small, then $T$ is invertible, which which means we can easily compute the density of $z$. 

(credit to random walks) 

Let us use the subscript notation $q_{[T]}$ to denote the distribution obtained by passing $q$ through $T$. Then we are interested in picking a $T$ which minimises $$\operatorname{KL}\left(q_{[T]} \| p\right)$$.

As this would minimize the discrepancy between q and p. Luckily we have built up a relationship on this topic that lets us use the kernalized Stien Discrepancy Giving the following solution

Theorem (Gradient of $\mathrm{KL}$ is the KSD) Let $x \sim q(x)$, and $T(x)=x+\epsilon \phi(x)$, where $\phi$ is a smooth function. Then

$$
\left.\nabla_\epsilon \mathrm{KL}\left(q_{[T]}|| p\right)\right|_{\epsilon=0}=-\mathbb{E}_{x \sim q}\left[\operatorname{trace} \mathcal{A}_p \phi(x)\right],
$$

where $q_{[T]}$ is the density of $T(x)$ and

$$
\mathcal{A}_p \phi(x)=\nabla_x \log p(x) \phi^{\top}(x)+\nabla_x \phi(x)
$$

## Expressing this concept numerically 

By construction we now need to identify a function satisfying $\|\phi\|_{\mathcal{H}^d} \leq 1$ 

For $f \in \mathcal{H}_D$ we have the following equality
$$
\begin{aligned}
\langle f, \beta\rangle_{\mathcal{H}_D} & =\sum_{d=1}^D\left\langle f_d(\cdot), \beta_d\right\rangle_{\mathcal{H}} \\
& =\sum_{d=1}^D\left\langle f_d(\cdot), \mathbb{E}_{x \sim q}\left[k(x, \cdot) \nabla_{x_d} \log p(x)+\nabla_{x_d} k(x, \cdot)\right]\right\rangle_{\mathcal{H}} \\
& \left.=\sum_{d=1}^D \mathbb{E}_{x \sim q}\left[\nabla_{x_d} \log p(x)\left\langle f_d(\cdot), k(x, \cdot)\right\rangle+\left\langle f_d(\cdot), \nabla_{x_d} k(x, \cdot)\right\rangle\right]\right\rangle_{\mathcal{H}} \\
& =\sum_{d=1}^D \mathbb{E}_{x \sim q}\left[\nabla_{x_d} \log p(x) f_d(x)+\nabla_{x_d} f_d(x)\right] \\
& =\mathbb{E}_{x \sim p}\left[\mathcal{A}_q f(x)\right]
\end{aligned}
$$

Therefore, the $f \in \mathcal{H}_D$ which maximises $\mathbb{E}_{x \sim p}\left[\mathcal{A}_q f(x)\right]$ is the one which maximises the inner product $\langle f, \beta\rangle_{\mathcal{H}_D}$, which occurs when $f$ is proportional to $\beta$.

If we normalize our final expression we obtain 

$$
\mathbb{E}_{x \sim q}\left[\nabla_{x_d} \log p(x) f_d(x)+\nabla_{x_d} f_d(x)\right]
$$

and for some kernel function k as our f in the hilbert space 

$$\mathbb{E}_{x \sim q}\left[k(x, \cdot) \nabla_x \log p(x)+\nabla_x k(x, \cdot)\right]$$


We now approximate $q$ by a finite set of $N$ particles at locations $x_n^{(i)}, n=1, \ldots, N$, at the $i^{\text {th }}$ iteration,since we cannot represent the entire function computationally. This yeilds 

$$ x_n^{(i+1)}=x_n^{(i)}+\frac{\epsilon^{(i)}}{N} \sum_{m=1}^N\left[\left.k\left(x_n, x_m\right) \nabla_x \log p(x)\right|_{x_n}+\left.\nabla_x k\left(x_m, x\right)\right|_{x_n}\right] .$$

Consider the RBF kernel 

$$k_{r b f}\left(\mathbf{x}, \mathbf{x}^{\prime}\right)=\exp -\frac{1}{2 \sigma^2}\left\|\mathbf{x}-\mathbf{x}^{\prime}\right\|^2$$

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
tf.compat.v1.disable_eager_execution()
```

```python
import tensorflow as tf


"""
svgd_kernel is an object that represents the svgd kernel function. 
    Input:
    -A nxd tensor "X" representing a set of n d dimensional points hoping to approximate the posterior distribution in question 
    Output: 
    -A nxn matrix Kxy the kernel matrix storing information about pairwise distance 
    -A nxd tensor dxkxy which is the gradient of Kxy with respect to X. This tensor contains parital derivatives of individual points in 
     the tensor with repsect to X
"""
def svgd_kernel(X):
    # Compute pairwise Euclidean distances between rows of X
    XY = tf.matmul(X, tf.transpose(X))
    X2_ = tf.reduce_sum(tf.square(X), axis=1)

    # Reshape X2_ to a column vector and tile it
    # to create a matrix X2e with the same dimensions as XY
    x2 = tf.reshape(X2_, shape=(tf.shape(X)[0], 1))
    X2e = tf.tile(x2, [1, tf.shape(X)[0]])

    # Compute the squared pairwise Euclidean distances
    H = tf.subtract(tf.add(X2e, tf.transpose(X2e)), 2 * XY)

    # Reshape H to a column vector
    V = tf.reshape(H, [-1, 1]) 

    # Compute the median distance between pairs of points in X
   
    def get_median(v):
        v = tf.reshape(v, [-1])
        m = v.get_shape()[0] // 2
        return tf.nn.top_k(v, m).values[m - 1]
    # Compute the bandwidth parameter for the RBF kernel
    h = get_median(V)
    h = tf.sqrt(0.5 * h / tf.math.log(tf.cast(tf.shape(X)[0], tf.float32) + 1.0))

    # Compute the RBF kernel and its gradient
    Kxy = tf.exp(-H / h ** 2 / 2.0)
    dxkxy = -tf.matmul(Kxy, X)
    sumkxy = tf.reduce_sum(Kxy, axis=1, keepdims=True)
    
    dxkxy = tf.add(dxkxy, tf.multiply(X, sumkxy)) / (h ** 2)

    return Kxy, dxkxy
```

```python
#Hyper parameter configuration
dtype = tf.float32
num_particles=128
num_latent = 2
lr = 0.1
alpha=.9
fudge_factor = 1e-6
num_iter = 300
range_limit = [-4,4]
npoints_plot=50
```

```python
def approximate_posterior(q):
    #Create a nxd tensor containing our particles that will approximate the posterior 
    z  = tf.compat.v1.placeholder(tf.float32, [num_particles, num_latent])
    
    # Compute Negative Energy 
    f = -q(z)
    
    prob = tf.exp(f)
    log_p_grad = tf.squeeze(tf.gradients(f, z))
    kernel_matrix, kernel_gradients = svgd_kernel(z) # call MAIN update

    grad_theta = (tf.matmul(kernel_matrix, log_p_grad) + kernel_gradients)/num_particles
    z_np = -np.random.randn(num_particles, num_latent)*2.0
    
    
#_______________________________________________Plotting 1_______________________________________________________________________ 
    fig, _ = plt.subplots(figsize=(20, 15))
    ax=plt.subplot(1,3,1, aspect='equal')
    plt.scatter(z_np[:,0], z_np[:,1])  
    plt.xlim(range_limit); plt.ylim(range_limit); ax.set_title('Initial particles: $p_0(z)$')
#______________________________________________________________________________________________________________________________
   
    # Here the session refers to svgd this loop iterates the particles updating with gradient information from svgd
    #num_iter times
    with tf.compat.v1.Session() as s:
        for i in range(num_iter):
            grad_theta_ = s.run( grad_theta, feed_dict={z: z_np } ) #compute gradient tensor containing information about all particles
            z_np = z_np + lr * grad_theta_ #KEY UPDATE STEP 

            
#_______________________________________________Plotting 2_______________________________________________________________________ 
    ax=plt.subplot(1,3,2, aspect='equal')
    plt.scatter(z_np[:,0], z_np[:,1])
    plt.xlim(range_limit); plt.ylim(range_limit); ax.set_title('Updated particles: $p_k(z)$')
#______________________________________________________________________________________________________________________________   
    
    
    z_pp  = tf.compat.v1.placeholder(tf.float32, [None, num_latent])
    prob = tf.exp(-q(z_pp))



#_______________________________________________Plotting 3_______________________________________________________________________ 
    ax=plt.subplot(1,3,3, aspect='equal')
    mesh_z1, mesh_z2, zv = evaluate_bivariate(range=range_limit, npoints=npoints_plot)
    with tf.compat.v1.Session() as s:
        phat_z = s.run(prob, feed_dict={z_pp: zv} ) 
    phat_z=phat_z.reshape([npoints_plot,npoints_plot])
    ax.pcolormesh(mesh_z1, mesh_z2, phat_z)
    z_min, z_max = -np.abs(phat_z).max(), np.abs(phat_z).max()
    plt.pcolor(mesh_z1, mesh_z2, phat_z, cmap='RdBu', vmin=z_min, vmax=z_max)
    plt.xlim(range_limit); plt.ylim(range_limit); ax.set_title('Target distribution: $q(z)$')
#______________________________________________________________________________________________________________________________
```

```python
def evaluate_bivariate(range, npoints):
    """Evaluate (possibly unnormalized) pdf over a meshgrid."""
    side = np.linspace(range[0], range[1], npoints)
    z1, z2 = np.meshgrid(side, side)
    zv = np.hstack([z1.reshape(-1, 1), z2.reshape(-1, 1)])
    
    return z1, z2, zv
```
